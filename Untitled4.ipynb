```python
import tensorflow as tf
if tf is not None:
    # Step 1: Create a random tensor of shape (4, 6)
    tensor = tf.random.uniform((4, 6))
    print("Original Tensor:", tensor.numpy())

    # Step 2: Find rank and shape
    rank = tf.rank(tensor)
    shape = tf.shape(tensor)
    print("Rank:", rank.numpy())
    print("Shape:", shape.numpy())

    # Step 3: Reshape into (2, 3, 4) and transpose to (3, 2, 4)
    reshaped_tensor = tf.reshape(tensor, (2, 3, 4))
    transposed_tensor = tf.transpose(reshaped_tensor, perm=[1, 0, 2])
    print("Reshaped Tensor Shape:", reshaped_tensor.shape)
    print("Transposed Tensor Shape:", transposed_tensor.shape)

    # Step 4: Broadcasting a smaller tensor (1, 4) to match (3, 2, 4) and adding
    small_tensor = tf.random.uniform((1, 4))
    broadcasted_tensor = tf.broadcast_to(small_tensor, (3, 2, 4))
    result_tensor = transposed_tensor + broadcasted_tensor
    print("Broadcasted Tensor Shape:", broadcasted_tensor.shape)
    print("Result Tensor Shape:", result_tensor.shape)

    # Step 5: Explanation of Broadcasting
    explanation = """
    Broadcasting in TensorFlow allows operations between tensors of different shapes by automatically expanding the smaller tensor
    along dimensions where its size is 1. In this case, the tensor of shape (1, 4) was expanded to match (3, 2, 4), enabling element-wise addition.
    """
    print(explanation)
else:
    print("Skipping TensorFlow operations due to missing module.")
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Step 1: Define true values (y_true) and model predictions (y_pred)
y_true = np.array([1, 0, 0])  # One-hot encoded class label
y_pred1 = np.array([0.8, 0.1, 0.1])  # Initial prediction
y_pred2 = np.array([0.7, 0.2, 0.1])  # Slightly modified prediction

# Step 2: Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE) losses
mse = tf.keras.losses.MeanSquaredError()
cce = tf.keras.losses.CategoricalCrossentropy()

mse_loss1 = mse(y_true, y_pred1).numpy()
cce_loss1 = cce(y_true, y_pred1).numpy()
mse_loss2 = mse(y_true, y_pred2).numpy()
cce_loss2 = cce(y_true, y_pred2).numpy()

# Print loss values
print(f"Initial Predictions - MSE: {mse_loss1}, CCE: {cce_loss1}")
print(f"Modified Predictions - MSE: {mse_loss2}, CCE: {cce_loss2}")

# Step 4: Plot loss function values using Matplotlib
labels = ['Initial MSE', 'Modified MSE', 'Initial CCE', 'Modified CCE']
values = [mse_loss1, mse_loss2, cce_loss1, cce_loss2]

plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['blue', 'blue', 'red', 'red'])
plt.ylabel('Loss Value')
plt.title('Comparison of MSE and Categorical Cross-Entropy Loss')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Step 1: Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values to range [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 2: Define the model architecture
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    return model

# Compile and train models using Adam and SGD
optimizers = {"Adam": tf.keras.optimizers.Adam(), "SGD": tf.keras.optimizers.SGD()}
history_dict = {}

for opt_name, optimizer in optimizers.items():
    print(f"Training with {opt_name} optimizer...")

    model = create_model()
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test), verbose=1)
    history_dict[opt_name] = history.history

# Step 3: Compare training and validation accuracy trends
plt.figure(figsize=(10, 5))

for opt_name, history in history_dict.items():
    plt.plot(history['val_accuracy'], label=f"{opt_name} - Validation")
    plt.plot(history['accuracy'], linestyle='dashed', label=f"{opt_name} - Training")

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Adam vs. SGD: Training and Validation Accuracy")
plt.legend()
plt.show()
```
